#4.Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?

Ans -

=VLOOKUP("E103",A2:C5,3,FALSE)

Will get as "Finance".

VLOOKUP function is used because the data is in vertical format and the lookup value (EmpID) is present in the first column. It helps to retrieve the corresponding department from the table using exact match.
